```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
class StresTestVerisi:
    def __init__(self, saniye=1000):
        self.saniye = saniye

    def veri_uret(self):
        donanim = {
            "CPU_Temp": np.random.randint(40, 100, size=self.saniye),
            "GPU_Temp": np.random.randint(40, 90, size=self.saniye),
            "Fan_Temp": np.random.randint(2000, 6000, size=self.saniye)
        }
        self.df = pd.DataFrame(donanim)
        return self.df
```

```python
class TermalAnalizML:
    def __init__(self, data):
        self.df = data
        self.scaler = StandardScaler()
        self.model = KMeans(
            n_clusters=3,
            random_state=42
        )

    def modeli_calistir(self):
            X_scaled = self.scaler.fit_transform(self.df)
            self.model.fit(X_scaled)
            predict = self.model.predict(X_scaled)
            self.df['Termal_Durum'] = predict
        
    def ozet_rapor(self):
            rapor = self.df.groupby('Termal_Durum').mean()
            return rapor
```

```python
# 1. Stres Testini Başlat (1000 saniyelik veri üret)
test_cihazi = StresTestVerisi(saniye=1000)
sensor_verileri = test_cihazi.veri_uret()

# 2. Makine Öğrenmesi ve EDA Sınıfına Gönder
analiz_motoru = TermalAnalizML(sensor_verileri)
analiz_motoru.modeli_calistir()

# 3. Raporu Çek ve Ekrana Bas
sonuc_tablosu = analiz_motoru.ozet_rapor()

print("🔍 TERMAL DURUM ÖZET RAPORU (KÜME ORTALAMALARI):")
print(sonuc_tablosu)
```
